```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
drive.mount('/content/drive')
path = "/content/drive/MyDrive/kidney_disease.csv"
```

```python
df = pd.read_csv(path)
df
```

```python
plt.bar(df["age"],df["sg"])
plt.title("SG")
plt.ylim(0.95,1.05)
plt.xlabel("Age")
plt.ylabel("SG")
plt.show()
```

```python
df = df.drop(columns = "appet")
df = df.drop(columns = "pe")
df = df.drop(columns = "ane")
df = df.drop(columns = "pcv")
df = df.drop(columns = "wc")
df = df.drop(columns = "rc")
df
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
df["rbc"] = df["rbc"].fillna("abnormal")
df["pc"] = df["pc"].fillna("abnormal")
df["pcc"] = df["pcc"].fillna("notpresent")
df["ba"] = df["ba"].fillna("notpresent")
df
```

```python
from pickle import SHORT_BINUNICODE
s_bgr = df["bgr"].sum()
s_mean = s_bgr/400
df["bgr"] = df["bgr"].fillna(s_mean)
s_bu = df["bu"].sum()
s_mean = s_bu/400
df["bu"] = df["bu"].fillna(s_mean)
s_sc = df["sc"].sum()
s_mean = s_sc/400
df["sc"] = df["sc"].fillna(s_mean)
s_sod = df["sod"].sum()
s_mean = s_sod/400
df["sod"] = df["sod"].fillna(s_mean)
s_pot = df["pot"].sum()
s_mean = s_pot/400
df["pot"] = df["pot"].fillna(s_mean)
s_hemo = df["hemo"].sum()
s_mean = s_hemo/400
df["hemo"] = df["hemo"].fillna(s_mean)
df
```

```python
plt.bar(df["age"],df["sg"])
plt.title("SG")
plt.ylim(0.95,1.05)
plt.xlabel("Age")
plt.ylabel("SG")
plt.show()
```
